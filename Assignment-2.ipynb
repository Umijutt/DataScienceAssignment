## pandas


```python
import pandas as pd
```

```python
data= pd.read_csv("Pokemon.csv")
data.head()
```

```python
data.sort_values(["HP","Speed"])
```

```python
data.tail(n=15)
```

```python
HP_mean = data["HP"].mean()

def set_HP(val):
    if val < HP_mean:
        return "HP Low"
    elif val == HP_mean:
        return "HP neutral"
    else:
        return "HP High"
```

```python
# "High_low_HP"  -> "HP Low", "HP High"
data["High_low_HP"] = data["HP"].apply(set_HP)
data
```

```python
speed_mean = data["Speed"].mean()

def set_speed(val):
    if val < speed_mean:
        return "Speed Low"
    elif val == speed_mean:
        return "Speed neutral"
    else:
        return "Speed High"
```

```python
# "Speed_high_low"  -> "Speed High", "Speed Low"
data["Speed_high_low"] = data["Speed"].apply(set_speed)
data
```
